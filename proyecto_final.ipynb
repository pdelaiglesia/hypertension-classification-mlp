```python
# Importar librerías
import pandas as pd
import sklearn
from sklearn import preprocessing
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler  
from sklearn.neural_network import MLPClassifier 
from sklearn.metrics import classification_report, confusion_matrix 
```

## Clasificación

```python
data = pd.read_csv("hipertension_preprocesado_final_balanceado.csv")
data
```

```python
# Takes first 10 columns and assign them to variable "X"
X = data.iloc[:, 0:10]
# Takes last column and assign them to variable "Y". 
y = data.iloc[:, 10] 
```

```python
X.head()
```

```python
y.head()
```

```python
# Now for train and test split (60% of  dataset into  training set and  other 40% into test data)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state=45)
```

```python
# Feature scaling
scaler = StandardScaler()  
scaler.fit(X_train)
X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test) 
```

```python
# Finally for the MLP- Multilayer Perceptron
mlp = MLPClassifier(hidden_layer_sizes=(15, 15), max_iter=2000) # 15 neuronas, 2 capas
mlp.fit(X_train, y_train.values.ravel())  
```

```python
predictions = mlp.predict(X_test)
```

```python
print(predictions)
```

```python
# Last thing: evaluation of algorithm performance in classifying flowers
print("Confusion matrix: ")
print(confusion_matrix(y_test,predictions))  
print(classification_report(y_test,predictions)) 
```
